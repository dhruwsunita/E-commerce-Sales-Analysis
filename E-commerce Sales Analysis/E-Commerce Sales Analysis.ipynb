### **Importing all the required Libraries** ###

```python
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = 'plotly_white'
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Sample - Superstore.csv', encoding='latin-1')
```

```python
# Converting date columns
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
# for shipping date
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.head()
```

```python
# Creating new columns
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Order Dat of Week'] = df['Order Date'].dt.day_of_week
```

```python
df.head()
```

## **Monthly Sales Analysis**

```python
sales_by_month = df.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
sales_by_month
```

```python
fig = px.line(sales_by_month,
              x='Order Month',
              y='Sales',
              title='Monthly Sales Analysis')
fig.show()
```

## **Sales By Category**

```python
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
fig = px.pie(sales_by_category,
             values='Sales',
             names='Category',
             hole=0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category', title_font=dict(size=24))

fig.show()
```

## **Sales by Sub-category**

```python
sales_by_sub_category = df.groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
sales_by_sub_category
```

```python
fig = px.bar(sales_by_sub_category,
             x='Sub-Category',
             y='Sales',
             title='Sales by Sub-category')
fig.show()
```

## **Profit by Category**

```python
profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()
```

```python
profit_by_category
```

```python
fig = px.pie(profit_by_category,
             values='Profit',
             names = 'Category',
             hole=0.5,
             color_discrete_sequence=colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Profit Analysis by Category', title_font=dict(size=24))

fig.show()
```

## **Profit by Sub-category**

```python
profit_by_sub_cat = df.groupby('Sub-Category')['Profit'].sum().reset_index()
profit_by_sub_cat
```

```python
fig = px.bar(profit_by_sub_cat,
             x='Sub-Category',
             y='Profit',
             color='Sub-Category',
             title='Profit by Sub-category')

fig.show()
```

## **Sales and Profit by Customer Segment**

```python
sales_profit_by_segment = df.groupby('Segment').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()

color_palette = colors.qualitative.Antique

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color = color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit',
                     marker_color = color_palette[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title = 'Customer Segment', yaxis_title='Amount')

fig.show()
```

## **Sales to Profit Ratio**

```python
sales_profit_by_segment  = df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
